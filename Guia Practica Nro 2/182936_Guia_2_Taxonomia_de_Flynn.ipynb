Ejercicio Propuesto: Crear un ejemplo que muestre una computación paralela de tipo MISD

```
#input:
#n=4
#procedimiento: 
#Instruccion1(n) duerme el proceso por "n" segundos y auemta el valor de "count"
#Instruccion2(n) reduce el valor de n en uno
#Instruccion3(n) calcula la suma de los "n" primeros naturales y reduce el valor de "n"
#output:
#imprime el tiempo actual y el valor de "count"
#imprime el tiempo actual, el numero de iteracion y el valor de "n"
#imprime el tiempo actual, el valor de "n" y la suma de los "n" primeros naturales
import time
import threading
n = 4
def Instruccion1(n): 
    count = 0
    print("Para la instruccion 1 en el momento: %s, su valor de n es: %s\n" % ( time.ctime(), n))
    while count < 3:
        time.sleep(n) #bloque de hilo 0 segundos, entoces se tiene una ejecución del hilo sin interrupciones
        count += 1
        print("instruccion 1 %s. count %s \n" % ( time.ctime(), count))

def Instruccion2(n):
    print("Para la instruccion 2 en el momento: %s, su valor de n es: %s\n" % ( time.ctime(), n)) 
    for i in range(n):
        print("instruccion 2 | iteracion: %s, en el momento: %s, su valor de n es: %s\n" % ( i, time.ctime(), n))
        n-=1
              
def Instruccion3(n):
    print("Para la instruccion 3 en el momento: %s, su valor de n es: %s\n" % ( time.ctime(), n))
    while 0 < n:
        print("instruccion 3 | suma de los %s, numeros naturales en el momento: %s, es: %s\n"%( n, time.ctime(),sum(range(n))))
        n-=1

        

#Creación de dos hilos que ejecutan la función print_time        
t1 = threading.Thread(target = Instruccion1, args = ( n, ))
t2 = threading.Thread(target = Instruccion2, args = ( n, ))
t3 = threading.Thread(target = Instruccion3, args = ( n, ))

#Inicialización de los hilos
t1.start()
t2.start()
t3.start()
```

En el ejemplo pódemos ver que existe la paralelizacion del tipo MISD porque existe un unico dato que vendria a ser el "n=4" que es utilizado por las tres instruciones que existen el cual cada una hace uso diferente del valor de n. Para poder ver todo lo que hace el ejemplo lo ejecute en python de mi computadora en donde podemos ver que en la priomera para de la ejecucion todas las instrucciones lo reacizan al mismo tiempo al igual que la instruccion 2 y 3 ya que no durmen el proceso y se ve que la instruccion 1 acaba ultipo ya que tiene que dormir el proceso cada 4 segundos =n=4, cada uno hace uso diferente del n.
